```python
from pickle import NONE
from numpy.core.numeric import full
import requests
from bs4 import BeautifulSoup

import mimetypes

def print_images(url, start=None, stop=None):
  response = requests.get('url')
  #created instamce of beautiful soup assigned to a variable
  soup = BeautifulSoup(response.text, 'html.parser')

  for img in soup.find_all('img')[start:stop]:
    #method starts with
    if img['src'].startswith('//'): #absolute path
    #made adjustment to full_img = https:// instead of https:
      full_img = 'https:' + img['src']
      #print(full_img)
    else:
        full_img = 'https://en.wikipedia.org' + img['src']
      #print(mimetype)
    mimetype = mimetypes.guess_type(full_img)
    if mimetype[0] is None:#if there is no index for for certain images
      continue#skip and continue to next iteration 
    if mimetype[0].startswith('image'):
        display(Image(full_img, width=100))
  
url = input('Give me url: ')
print_images(url)
```
